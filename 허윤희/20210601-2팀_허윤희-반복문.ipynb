#**반복문**
- for문 : 횟수 제어
- while문 : 조건에 따라 변경!!! 무한루프에 빠질 수 있다. 

## for
- 주어진 자료를 순회할때 주로 사용되는 반복구조

```python
for i in [1,2,3,4,5,6,7]:
    print(i)
```

```python
for i in 'python data':
    print(i)
```

# **범위 > range() 함수**

- 시작값/증가값 생략 가능
- 끝값 생략 불가

```python
for i in range(1,6):
    print(i)
```

```python
# 범위값
for i in range(5):
    print(i,end= " ") # end='' 줄이 바뀌지 않고 출력된다
```

```python
# 시작값, 범위값
for i in range(1,6):
    print(i,end= "  ") # 공백을 지정하면 공백만큼의 간격을 두고 출력한다
```

```python
# 증가값
for i in range(1,10,2):
    print(i,end= "\t")  # \t 탭 간격으로 출력된다.
```

```python
#-역순 : 오류(범위 밖) _ 작은 수부터 큰 수까지 -1? (정정: 10부터 1까지 -1 해라)
for i in range(1,10,-1):
    print(i,end= " ")
```

```python
#역순 : 올바른 예시
for i in range(10,0,-1):
    print(i,end= " ")
```

```python
for i in range(100):
    print('Hello World!!')
```

## **while**
- 조건식의 결과가 True 동안 반복 / False가 되면 반복문을 빠져나옴
- while 조건식:
    명령블럭



```python
i = 0 
#조건식에 사용되는 변수는 while문에 들어가기 전에 초기화 되어야 한다.

while i <= 30:
    print(i,end= ' ')
    i = i + 3
```

```python
for i in range(0,31,3):
    print(i,end=' ')
```

무한루프일때 `ctrl+ m i`

```python
#무한루프 
i = 1 
while i <= 5: # i값이 계속 1이므로 무한루프
    print(i)
print('end')
```

### 예제 3

```python
# n개의 정수를 입력 받아 짝수의 합계 구하기
sum1 = 0
n = int(input('몇개의 숫자를 더할까요? :'))
for i in range(n):
    num = int(input('정수입력:'))
    if num % 2 == 0:  # 짝수 조건
        sum1 += num
print('합계출력:',sum1)
```

```python
# n개의 정수를 입력 받아 3의 배수의 합계 구하기
sum1 = 0
n = int(input('몇개의 숫자를 더할까요? :'))
for i in range(n):
    num = int(input('정수입력:'))
    if num % 3 == 0:  # 3의 배수 조건
        sum1 += num
print('합계출력:',sum1)
```

### 예제 > 1 ~ n 까지 더하는 프로그램

```python
sum_1 = 0
n = int(input('1부터 몇까지 더할까요? :'))
for i in range(1,n+1):
    sum_1 += i
print(f'총 합계는 {sum_1}입니다')
```

### 예제 > n개 숫자의 평균 구하기

```python
n = int(input('입력받을 정수의 개수 :'))
sum_1 = 0

for i in range(n):
    num = int(input('숫자를 입력하세요 :'))
    sum_1 += num
if n == 0:
    print('0개의 정수를 입력하셨습니다.')
else:
    print('평균', sum_1/n)
```

```python
#임의의 개수 숫자의 평균
n = int(input('입력받을 정수의 개수: '))
num_list = []
for i in range(n):
    num_list.append(int(input('정수: ')))
    sum0 = sum(num_list)
print(f'평균: {sum0/n}')
```

### 예제 > 팩토리얼 계산하기


```python
n = int(input('몇 팩토리얼? :'))
factorial = 1
for i in range (n,0,-1):
    factorial *= i
print(f'{n}!={factorial}')
```

### 예제 > up & down 게임
1. 1~100 사이의 정답 숫자를 랜덤으로 하나 정하고
2. 정답 숫자를 맞출 때까지 숫자를 입력하는 게임.
3. 입력된 숫자가 정답보다 작으면 'Down', 크면 'Up'을 출력하고
4. 숫자를 다시 입력받는다.
5. 정답을 맞추면 정답!이라고 출력하고 게임을 끝낸다

```python
import random
answer = random.randint(1,100)
num = 0
while num != answer:
    num = int(input('숫자를 입력하세요: '))
    if num > answer:
        print('Down')
    elif num < answer:
        print('Up')
print(f'정답! 답은 {answer}입니다!')
```
